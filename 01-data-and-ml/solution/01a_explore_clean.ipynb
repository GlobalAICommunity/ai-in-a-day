# 1a · Explore & clean the colony telemetry — SOLUTION

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../../data/telemetry.csv", parse_dates=["timestamp"])
print(df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isna().sum()
```

```python
df = df.ffill().bfill()
assert df.isna().sum().sum() == 0
print("No missing values remain.")
```

```python
fig, axes = plt.subplots(3, 1, figsize=(11, 8), sharex=True)
axes[0].plot(df["timestamp"], df["o2_pct"], color="tab:blue")
axes[0].axhspan(19.5, 23.5, color="green", alpha=0.08)
axes[0].set_ylabel("O2 %")
axes[1].plot(df["timestamp"], df["co2_ppm"], color="tab:red")
axes[1].axhspan(400, 1000, color="green", alpha=0.08)
axes[1].set_ylabel("CO2 ppm")
axes[2].plot(df["timestamp"], df["power_kw"], color="tab:orange")
axes[2].set_ylabel("Power kW")
axes[0].set_title("Orbital Module B — one week of telemetry")
plt.tight_layout()
plt.show()
```

```python
# The O2 drop / CO2 rise is the scrubber fault on Day 5 (~14:00).
# Highlight labeled incidents on the O2 chart:
plt.figure(figsize=(11, 3))
plt.plot(df["timestamp"], df["o2_pct"], color="lightgray")
inc = df[df["label"] == 1]
plt.scatter(inc["timestamp"], inc["o2_pct"], color="red", s=8, label="labeled incident")
plt.legend(); plt.title("Known incidents"); plt.show()

df[["o2_pct", "co2_ppm", "power_kw", "battery_pct", "dust_index"]].corr()
```
