## Model selection: big vs small

On a power-constrained colony, a smaller model that runs faster and cooler can be worth a
small quality trade-off. Compare `llama3.2` with `phi3.5`.

```python
import time
from shared.llm import CHAT_MODEL, SMALL_MODEL

question = "Explain in 2 sentences why CO2 buildup is dangerous in a sealed habitat."

for model in (CHAT_MODEL, SMALL_MODEL):
    t0 = time.time()
    answer = chat(question, model=model, temperature=0.2)
    print(f"\n=== {model}  ({time.time()-t0:.1f}s) ===\n{answer}")
```
