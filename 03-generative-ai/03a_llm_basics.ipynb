## Step 2 — Temperature: focused vs creative

**Temperature** controls how much randomness the model uses when picking its next token.

- **Low (e.g. 0.0)** → the model almost always picks the most likely next word. Answers are
  focused, consistent, and repeatable. Good for facts, code, and anything where you want the
  same answer every time.
- **High (e.g. 1.2)** → the model takes more chances. Answers are more varied and creative —
  and more likely to wander. Good for brainstorming, bad for precision.

Run the cell a few times at each setting and watch how the *low*-temperature answer barely
changes while the *high*-temperature one keeps surprising you.

## Step 3 — Choosing a model: big vs small

Bigger models are generally more capable but slower and hungrier for memory and power. On a
colony where every watt counts, a smaller model that answers *well enough* in half the time
can be the smarter choice. This is a genuine engineering trade-off you'll face in real
projects too (cost, latency, privacy, battery).

Below we ask both models the same question and time how long each takes. `CHAT_MODEL` is
`llama3.2` (the bigger one) and `SMALL_MODEL` is `phi3.5` (the smaller one). Compare both the
**quality** of the answers and the **seconds** they took.

```python
import time
from shared.llm import CHAT_MODEL, SMALL_MODEL

question = "Explain in 2 sentences why CO2 buildup is dangerous in a sealed habitat."

for model in (CHAT_MODEL, SMALL_MODEL):
    t0 = time.time()
    answer = chat(question, model=model, temperature=0.2)
    print(f"\n=== {model}  (took {time.time()-t0:.1f}s) ===\n{answer}")
```
