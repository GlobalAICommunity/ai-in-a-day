```python
import sys, os, time
sys.path.insert(0, os.path.abspath("../.."))
from shared.llm import chat, CHAT_MODEL, SMALL_MODEL

print(chat("In one sentence, what is a Mars research colony?"))
```

```python
prompt = "Give the colony a one-line morning greeting."
print("temp 0.0:", chat(prompt, temperature=0.0))
print("temp 1.2:", chat(prompt, temperature=1.2))
```

```python
question = "Explain in 2 sentences why CO2 buildup is dangerous in a sealed habitat."
for model in (CHAT_MODEL, SMALL_MODEL):
    t0 = time.time()
    answer = chat(question, model=model, temperature=0.2)
    print(f"\n=== {model}  ({time.time()-t0:.1f}s) ===\n{answer}")
```
