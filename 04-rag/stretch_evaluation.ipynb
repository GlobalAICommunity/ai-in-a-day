# 🌱 Stretch · Evaluating RAG

**Self-paced**

"It looked right" is not a metric. Here we measure whether retrieval actually pulls the
*correct manual* for a set of known questions — a simple **retrieval hit-rate**.

This is the beginner's on-ramp to LLM evaluation: build a tiny labeled set, run it, score it.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from shared.rag import RagIndex

index = RagIndex.from_manuals()

# A small gold set: question -> the manual we expect to be retrieved.
gold = [
    ("What do I do if oxygen drops?", "emergency-procedures.md"),
    ("How do we handle a dust storm?", "power-systems.md"),
    ("What is the pre-EVA checklist?", "eva-safety.md"),
    ("How do I recover a failed CO2 scrubber?", "life-support.md"),
    ("Can we wait for Mission Control approval in an emergency?", "comms.md"),
]
```

```python
hits_at_3 = 0
for question, expected in gold:
    sources = {h["source"] for h in index.retrieve(question, k=3)}
    ok = expected in sources
    hits_at_3 += ok
    print(f"{'✅' if ok else '❌'}  {question}  ->  {sorted(sources)}")

print(f"\nHit-rate@3: {hits_at_3}/{len(gold)} = {hits_at_3/len(gold):.0%}")
```

### Try this

- Lower `k` to 1 — does the hit-rate drop?
- Add trickier, differently-worded questions to `gold`.
- **Groundedness** is the next metric: does the *answer* only use retrieved text? You can
  even use the LLM itself as a judge ("LLM-as-judge").
