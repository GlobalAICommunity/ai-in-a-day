**Core · Live-code · ~25 min**

This is the moment ARIA gains hands. We'll give it a set of **tools** (functions it can ask us
to run) and watch it decide, on its own, which to call to answer a question. That decision-
making loop is what makes it an **agent** rather than a chatbot.

**Needs:** Ollama with `llama3.2` (which supports tool calling). Solution:
[`solution/05a_tool_use.ipynb`](solution/05a_tool_use.ipynb).

## Step 1 — What the model actually sees

For the model to use a tool, we have to *describe* it in a structured way: its name, a
human-readable description, and the arguments it accepts (with types). This description is
called a **tool schema**. The model reads these schemas and uses the descriptions to decide
which tool fits the question and what arguments to pass.

Let's print one schema so it's concrete. Notice it's just structured data describing a
function — nothing magic.

```python
import sys, os, json
sys.path.insert(0, os.path.abspath(".."))
from shared.agent import run_agent
from shared.tools import TOOLS, read_only_tools

# The schema for the get_telemetry tool -- this is exactly what the model is shown.
print(json.dumps(TOOLS["get_telemetry"]["schema"], indent=2))
```

## Step 2 — Let ARIA use a tool

`run_agent(system, user, tools)` runs the full agent loop we described in the README: it sends
the question plus the tool schemas to the model, and if the model asks to call a tool, it runs
the function and feeds the result back — looping until ARIA gives a final answer.

We give ARIA only the **read-only** tools here (`get_telemetry`, `search_manual`) — the safe
ones that can't change anything. By default `run_agent` prints each tool call, so you can
literally watch ARIA decide to check the oxygen sensor before answering.
