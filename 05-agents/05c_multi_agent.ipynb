# 5c · A team of agents (multi-agent)

**Core · Live-code · ~20 min**

Sometimes one assistant trying to do everything gives muddled results. A common, powerful
pattern is to use **several specialised agents** and have them collaborate — just like a real
crew. We'll build three: an **Engineer**, a **Medic**, and a **Commander** who weighs their
advice and makes the call.

Solution: [`solution/05c_multi_agent.ipynb`](solution/05c_multi_agent.ipynb).

## Why multiple agents?

Three practical reasons:
1. **Focus** — each agent gets a short, specific system prompt, so it reasons better about its
   own domain than one giant do-everything prompt would.
2. **Different tools/permissions** — the Engineer might read sensors; the Medic might not need
   tools at all. You can grant each agent exactly what it needs (least privilege again).
3. **A clear decision-maker** — a **coordinator** agent (here, the Commander) gathers the
   specialists' input and produces one decision, avoiding contradictory advice.

Each agent is just another `run_agent(...)` call with a different system prompt. That's the
beauty of it — no new machinery, just roles.
