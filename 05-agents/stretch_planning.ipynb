# 🌱 Stretch · The planning pattern

**Self-paced**

Instead of reacting step-by-step, a planning agent first writes a **plan**, then executes
it. This makes its behaviour easier to review — important when actions have consequences.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from shared.llm import chat
from shared.agent import run_agent
from shared.tools import read_only_tools

GOAL = "Assess whether Module B is safe to occupy right now and recommend next steps."
```

```python
# Step 1: make a plan
plan = chat(
    f"Goal: {GOAL}\nList 3-5 concise numbered steps to achieve it. Steps only.",
    system="You are ARIA, a careful planner for the Orbital colony.",
    temperature=0,
)
print("PLAN:\n", plan)
```

```python
# Step 2: execute the plan with tools
answer = run_agent(
    "You are ARIA. Follow the given plan, using tools to gather real data, then give a "
    "final safety assessment. Physical actions need human confirmation.",
    f"Goal: {GOAL}\n\nPlan:\n{plan}",
    read_only_tools(),
)
print("\nRESULT:\n", answer)
```

### Why separate planning from acting?

- A human can review the plan **before** anything happens.
- The plan gives the agent structure, reducing aimless tool calls.
- For high-stakes systems, "plan → approve → execute" is a safer pattern than free-running.
